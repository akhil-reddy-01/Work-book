```python
import torch
import torch.nn as nn
import torch.nn.functional as f
from torch.utils.data import DataLoader
from torch.utils.data import TensorDataset
```

```python
batchsize=32
n=100    #epochs=100
lr=0.0005
```

```python
def dataload(X,Y):
  X=torch.from_numpy(X).float().unsqueeze(-1)
  Y=torch.from_numpy(Y).float().unsqueeze(-1)
  dataset = TensorDataset(X, Y)
  traindataloader = DataLoader(dataset, batchsize, shuffle=True)
  return traindataloader
```

```python
def fit(n, model, losfun, optmsr, train_dl):
    loss_list=[]
    #n=number of epochs
    for epoch in range(n):
        
        for xb,yb in train_dl:
            
            pred = model(xb)
            loss = losfun(pred, yb)
            loss_list.append(loss)
            #Compute gradients
            loss.backward()
            
            #Updating parameters using gradients
            optmsr.step()
            
            #Reset the gradients to zero
            optmsr.zero_grad()

        
        if (epoch) != 0:
            print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, n, loss.item()))
    
    bb = torch.FloatTensor(loss_list)
    plt.plot(bb, 'r')
    plt.show()
```

```python
WandB=[]
Alosstest=[]
Alosstrain=[]
for aa in range(4):
  for ba in range(4):
    if (aa!=ba):
      x=train.iloc[:,aa].values
      y=train.iloc[:,ba].values
      xt=test.iloc[:,aa].values
      yt=test.iloc[:,ba].values
      train_dl=dataload(x,y)
      test_dl=dataload(xt,yt)
      model = nn.Linear(1,1)
      losfun=f.mse_loss
      optmsr = torch.optim.SGD(model.parameters(), lr)
      fit(n, model, losfun, optmsr, train_dl)
      
      for x,y in test_dl:
        preds = model(x)
        print("Prediction is :n",preds)
        #print("nActual targets is :n",y)
        losstest=losfun(preds, y).item()
        print("nLoss is: ",losstest)
        break
      WandB.append([model.weight, model.bias])
      Alosstest.append(losstest)
```

```python
def fitt(n, model, losfun, optmsr, train_dl):
    loss_list=[]
    #n=number of epochs
    for epoch in range(n):
        
        for xb,yb in train_dl:
            
            pred = model(xb)
            loss = losfun(pred, yb)
            l2_lambda = 0.001
            l2_norm = sum(p.pow(2.0).sum() for p in model.parameters())

            loss = loss + l2_lambda * l2_norm

            
            loss_list.append(loss)
            #Compute gradients
            loss.backward()
            
            #Updating parameters using gradients
            optmsr.step()
            
            #Reset the gradients to zero
            optmsr.zero_grad()

        
        if (epoch) != 0:
            print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, n, loss.item()))
    
    bb = torch.FloatTensor(loss_list)
    plt.plot(bb, 'r')
    plt.show()
```
